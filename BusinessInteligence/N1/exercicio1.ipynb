```python
import statistics

def addNum():
    priBin = float(input("Digite a nota do 1° bimestre: "))
    segBin = float(input("Digite a nota do 2° bimestre: "))
    terBin = float(input("Digite a nota do 3° bimestre: "))
    quaBin = float(input("Digite a nota do 4° bimestre: "))
    
    entradas = [priBin, segBin, terBin, quaBin]
    
    media = statistics.mean(list(entradas))
    if media <5:
        print("\nO estudante está em recuperação, pois sua média foi,", media)
    else:
        print("\nO estudante foi aprovado com média, ", media)
```
